```python
# Distribution Plots
# distplot, joinplot, pairplot
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
# Correlation with Heatmap
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
# Join Plot
# Bivariate Analysis
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

```python
# Pair Plot -> Scatter plot (more features)

sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='smoker')
```

```python
df['smoker'].value_counts()
```

```python
# Dist Plot
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

```python
# Categorical Plots
# Countplot(), Violinplot(), Boxplot(), Barplot()

# Count Plot
sns.countplot('sex', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

```python
# Bar Plot
sns.barplot(x='total_bill', y='sex', data=df)
```

```python
sns.barplot(y='total_bill', x='sex', data=df)
```

```python
# Box Plot
sns.boxplot('smoker', 'total_bill', data=df)
```

```python
sns.boxplot('day', 'total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

```python
# Violin Plot (Combination of box plot and KDE)
sns.violinplot(x='total_bill', y='day', data=df, paletter='rainbow') 
```
