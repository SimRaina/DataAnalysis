```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
# Scatter plot
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
# Plot
plt.plot(x,y,'r*--', linestyle='dashed', linewidth=2, markersize=12)
```

```python
plt.plot(x,y,'ro--', linestyle='dashed', linewidth=2, markersize=12)
```

```python
plt.plot(x,y,'r+--', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
# Sub Plots
plt.subplot(2,2,1) # First figure
plt.plot(x,y,'r--')
plt.subplot(2,2,2) # Second figure
plt.plot(x,y,'g*--')
plt.subplot(2,2,3) # Third figure
plt.plot(x,y,'bo--')
plt.subplot(2,2,4) # Fourth figure
plt.plot(x,y,'go')
```

```python
# Line plot
x=np.arange(1,11)
y=3*x+5
plt.title('Matplotlib Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4* np.pi, 0.1)
y=np.sin(x)
plt.title('Sine Wave Form')
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4* np.pi, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

plt.show()
```

```python
# Bar Plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y, align='center')
plt.bar(x2,y2, color='g')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
# Histogram
a=np.array([22,87,5,43,56,88,77,34,56,76,54,31,27])
plt.hist(a, bins=20)
plt.title('Histogram')
plt.show()
```

```python
# Box Plot
data=[np.random.normal(0, std, 10) for std in range(1,4)]
print(data)
plt.boxplot(data, vert=True, patch_artist=True);
```

```python
# Pie Chart
labels='Python', 'Java', 'Ruby', 'C++'
sizes=[215, 130, 245, 210]
colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode=(0.1, 0, 0, 0) # Explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
