```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green')
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=3,alpha=0.5) # RGB Hex Codes
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',lw=3,alpha=0.5) # RGB Hex Codes
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',lw=0.5,ls='-',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
# '-.', ':','--'  linestyle=ls
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
