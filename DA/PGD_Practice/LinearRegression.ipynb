```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
X, Y = make_regression(n_features = 1, noise = 10, n_samples = 100)
```

```python
plt.xlabel('Independent Variable = X')
plt.ylabel('Dependent Variable = Y')
plt.scatter(X, Y, s=3)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,Y)
```

```python
lr.coef_, lr.intercept_
```

```python
pred = lr.predict(X)
```

```python
plt.scatter(X, Y, s=1, label = 'training')
plt.scatter(X, pred, s=1, label = 'prediction')
plt.xlabel('Independent Variable - X')
plt.ylabel('Dependent Variable - Y')
plt.legend()
plt.show()
```
