```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
all_tables = pd.read_html('https://fbref.com/en/comps/82/Indian-Super-League-Stats')
```

```python
team_stats = pd.DataFrame(all_tables[3])
```

```python
team_stats.columns = [' '.join(col).strip() for col in team_stats.columns]
```

```python
team_stats.columns
```

```python
team_stats.rename(columns={team_stats.columns[3]: 'Poss'}, inplace=True)
team_stats.rename(columns={team_stats.columns[0]: 'Team'}, inplace=True)
```

```python
team_stats.columns
```

```python
# Question1: Get the team with the highest possession
top_possession_team = team_stats[['Team', 'Poss']].sort_values(by='Poss', ascending=False).head(1)
print(top_possession_team)
```
